## 네이버 뉴스 섹션 메뉴와 섹션별 url 추출

```python
from urllib.request import urlopen
import bs4
import pandas as pd
```

```python
url = 'https://news.naver.com'
html = urlopen(url)
bs_obj = bs4.BeautifulSoup(html, 'html.parser')
```

```python
# 네이버 뉴스 섹션메뉴 태그 확인 (개발자도구)
#body > section > header > div.Nlnb._float_lnb > div > div > div.Nlnb_left._lnb_scroll > div > div > ul
#select가 너무 길어서 유용하지 않음
# 직접 확인한 태그의 클래스와 속성 사용
# ul .Nlnb_menu_list
uls = bs_obj.findAll('ul',{'class' : 'Nlnb_menu_list' })
len(uls)
ul = bs_obj.find('ul',{'class' : 'Nlnb_menu_list' })
ul
```

```python
lis = ul.findAll('li')

len(lis)
```

```python
for li in lis :
    a_tag = li.find('a')
    print(a_tag.text, ';', a_tag['href'])
```

```python
# li 태그를 이용해서 추출 
lis = bs_obj.findAll('li',{'class':'Nlist_item'})
len(lis)
```

```python
# 수집 데이터 df 구성
section = []
link = []
```

```python
for li in lis :
    a_tag = li.find('a')
    section.append(a_tag.text)
    link.append(a_tag['href'])
```

```python
col_dict = {'section' : section, 'link':link}
```

```python
news_section_df = pd.DataFrame(col_dict)
```

```python
news_section_df
```

```python
news_section_df.to_csv('./crawl_data/naver_news_section.csv')
```
