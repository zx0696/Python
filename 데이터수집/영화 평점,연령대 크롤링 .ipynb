```python
driver = webdriver.Chrome('/driver/chromedriver.exe')
base_url = 'https://movie.daum.net/search?q='
search_keyword = movie_title
years_list = []
relation_date_list = []
grade_list = []

for n in search_keyword :
    if not isinstance(n, str):
        continue
    url = base_url + n
    driver.get(url)
    sel = '#mainContent > div > div.box_searchinfo > div.detail_searchinfo.movie.search_result > ul > li > div > div > strong > a'
    elems = driver.find_elements(By.CSS_SELECTOR, sel)
    for i in range(0,len(elems)) :
        try :
            if elems[i].text == n :
                elems[i].click()
        except : pass
    time.sleep(2)
    #개봉일 크롤링 
    sel_rd = '#mainContent > div > div.box_basic > div.info_detail > div.detail_cont > div:nth-child(1) > dl'
    elems_rd = driver.find_elements(By.CSS_SELECTOR, sel_rd)
    try :
        for i in range(0,len(elems_rd )) :
            if elems_rd [i].text.startswith('개봉')== True :
                relation_date_list.append(elems_rd [i].text)
    except : pass

    # 관람등급 크롤링
    sel_years = '#mainContent > div > div.box_basic > div.info_detail > div.detail_cont > div:nth-child(1) > dl'
    elems_years = driver.find_elements(By.CSS_SELECTOR, sel_years)
    try :
        for i in range(0,len(elems_years )) :
            if elems_years [i].text.startswith('등급')== True :
                years_list.append(elems_years [i].text)
    except : pass

    # 관람평점 크롤링
    sel_grade='#mainContent > div > div.box_basic > div.info_detail > div.detail_cont > div:nth-child(2) > dl'
    elems_grade = driver.find_elements(By.CSS_SELECTOR, sel_grade)
    try :
        for i in range(0,len(elems_grade )) :
            if elems_grade [i].text.startswith('평점')== True :
                grade_list.append(elems_grade [i].text)
    except : pass 
```

```python
# 영화정보
base_url = 'https://movie.daum.net/search?q='
search_keyword = movie_info

movie_info_list = pd.DataFrame(columns=['제목','개봉','관람등급','평점'])

for n in search_keyword :
    url = base_url + str(n)
    driver.get(url)
    time.sleep(1)
    sel = '#mainContent > div > div.box_searchinfo > div.detail_searchinfo.movie.search_result > ul > li > div > div > strong > a'
    elems = driver.find_elements(By.CSS_SELECTOR, sel)
    
    for i in range(0,len(elems)) :
        try :
            if elems[i].text == n :
                elems[i].click()
        except : pass
    time.sleep(2)
    sel_info = '#mainContent > div > div.box_basic > div.info_detail > div.detail_cont > div > dl'
    elems_info = driver.find_elements(By.CSS_SELECTOR, sel_info)
#     sel_tit = '#mainContent > div > div.box_basic > div.info_detail > div.detail_tit > h3 > span.txt_tit'
#     elem_tit = driver.find_element(By.CSS_SELECTOR, sel_tit)
    Title = n
    for i in range(0,len(elems_info)) :
        if elems_info[i].text.startswith('개봉') == True :
            relationDt = elems_info[i].text
        elif elems_info[i].text.startswith('등급') == True :
            ageRt = elems_info[i].text
        elif elems_info[i].text.startswith('평점') == True :
            grade = elems_info[i].text
    movie_info_list = movie_info_list.append({'제목':Title,'개봉':relationDt,'관람등급':ageRt,'평점':grade},ignore_index=True)
```
