### 연습문제 1
- 두개의 데이터프레임을 만들고 merge 명령을 병합한다.
- 단, 데이터 프레임은 다음 조건을 만족해야 한다.
    1. 각각 5X5 이상의 크기를 가진다.
    2. 공통열을 하나이상 갖는다. 
        - 다만 공통열의 이름은 서로 다르게 할 것
    3. merge의 경우를 inner, outer, left, right 4개의 형태로 출력할 것
    4. 지정된 인덱스와 컬럼명을 갖는다. 

```python
data = [[22, 60.1, 170.5, '남', '서울'], 
        [45, 51.3, 157.3, '여', '부산'], 
        [22, 68.3,180.1,  '남', '대구'],
        [33, 88.3, 190.2, '남', '제주'], 
        [27, 48.3, 160.1, '여', '강릉']]
df1 = pd.DataFrame(data, 
                   index=['홍길동', '이몽룡', '성춘향','변학도','김연아'], 
                   columns=["나이", "몸무게", "키", "성별" ,"주소"])
df1
```

```python
data = [[22, 60.1, 170.5, '남', '서울'], 
        [45, 51.3, 157.3, '여', '부산'], 
        [22, 68.3,180.1,  '남', '울산'],
        [33, 88.3, 190.2, '남', '제주'], 
        [27, 48.3, 160.1, '여', '광주']]
df2 = pd.DataFrame(data,
                   index=[1,2,3,4,5],
                   columns=["나이", "몸무게", "키", "성별" ,"지역"])
df2
```

```python
df1.columns
df2.columns
# 병합을 진행할 때 키가 되는 조건은 행을 유니크하게 설정할 수 있는 컬럼
# 실세계에서 구분하기 위해 명령하는 개념
```

```python
pd.merge(df1,df2, left_on = '주소', right_on='지역',how='outer')
pd.merge(df1,df2, left_on = '주소', right_on='지역',how='inner') # 기본값
pd.merge(df1,df2, left_on = '주소', right_on='지역',how='left')
pd.merge(df1,df2, left_on = '주소', right_on='지역',how='right')
pd.merge(df1,df2, on='나이')
```

### 연습문제 2
- 아래 csv 파일을 다운 받아 데이터프레임을 생성하고 인덱싱을 통해 일부를 추출 해 문제에 사용할 데이터 프레임을 추출한다
- 추출된 데이터 프레임을 사용해 제시된 문제를 해결한다

Dataurl = ‘https://raw.githubusercontent.com/Datamanim/pandas/main/mergeTEst.csv’

```python
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/mergeTEst.csv',index_col= 0)
df
```

```python
df1 = df.iloc[:5,:]
df2 = df.iloc[4:,:]

display(df1)
display(df2)
```

```python
# df1과 df2 데이터를 하나의 데이터 프레임으로 합치세요  # axis= 행방향 결합은 0  열방향은 1 
total = pd.concat([df1,df2])
total
```

```python
# df1과 df2를 결합하되 중복행은 한번만 표시되게 결합하시오- 중복열은 허용 함
total =pd.concat([df1,df2],axis=1)
total
```

```python
df3 = df.iloc[:2,:4]
df4 = df.iloc[5:,3:]

display(df3)
display(df4)
```

```python
# df3과 df4 데이터를 하나의 데이터 프레임으로 결합하되 둘다 포함하고 있는 년도에 대해서만 작업을 진행하세요
pd.concat([df3,df4], join='inner')
```

```python
# df3과 df4 데이터를 하나의 데이터 프레임으로 합치시오. 모든 컬럼을 포함하고, 결측치는 0으로 대체하세요
pd.concat([df3,df4]).fillna(0).astype(int)
```

```python
df.T # 행과 열을 바꿔주는 전치
```

```python
df5 = df.T.iloc[:7,:3]
df6 = df.T.iloc[6:,2:5]

display(df5)
display(df6)
```

```python
# df5과 df6 데이터를 하나의 데이터 프레임으로 merge함수를 이용하여 합치시오 
# Algeria컬럼을 key로 하고 두 데이터 모두 포함하는 데이터만 출력하세요
df5.merge(df6, on='Algeria')
```

```python
# df5과 df6 데이터를 하나의 데이터 프레임으로 merge함수를 이용하여 합치시오
# Algeria컬럼을 key로 하고 합집합으로 합치세요
df5.merge(df6, on='Algeria',how='outer')
```

```python
# df5, df6 데이터 프레임을 인덱스 활용해서 병합하시오 
pd.merge(df5, df6, how='inner', left_index=True,right_index=True)
```

### 연습문제 3
- 어느회사의 전반기(1월-6월)실적을 나타내는 데이터프레임과 후반기(7월-12월)실적을 나타내는  데이터 프레임을 작성한다

- 실적 정보는 "매출","비용","이익"으로 이루어진다.
- '매출','비용'은 임의의 데이터를  사용하고
- 이익은 이익=매출-비용의  가공필드로 생성한다
    
- 또한 1년간의 총 실적을 마지막 열로 덧붙인다.

```python
# 데이터프레임 생성

month = [str(i)+'월' for i in range(1,13)]
month
```

```python

np.random.seed(3)

#매출, 비용 data random  생성
data={}
for i in month[0:6] :
    data[i] = np.random.randint(1,100,2)
data 
```

```python
#데이터프레임 생성
df_score3 = pd.DataFrame(data,
                        index=['매출','비용'])
display(df_score3)
```

```python
#가공 행 이익 생성 코드 작성
df_score3.loc['이익'] = df_score3.loc['매출'] - df_score3.loc['비용']
df_score3
```

```python

display(df_score3)

#매출, 비용 data random  생성
data={}
for i in month[6:12] :
    data[i] = np.random.randint(1,100,2)
data 

#데이터프레임 생성
df_score4 = pd.DataFrame(data,
                        index=['매출','비용'])
display(df_score4)

#가공 행 이익 생성 코드 작성
#가공 행 이익 생성 코드 작성
df_score4.loc['이익'] = df_score4.loc['매출'] - df_score4.loc['비용']
df_score4

df_score4
```
