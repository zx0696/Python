```python
driver = webdriver.Chrome('./driver/chromedriver.exe')
```

```python
base_url = 'https://movie.daum.net/search?q='
search_keyword = '히말라야'
```

```python
url = base_url + search_keyword
```

```python
driver.get(url)
```

```python
sel = '#mainContent > div > div.box_searchinfo > div.detail_searchinfo.movie.search_result > ul > li > div > div > strong > a'
elems = driver.find_elements(By.CSS_SELECTOR, sel)
```

```python
elems[1].click()
```

```python
for i in range(0,len(elems)) :
    try :
        if elems[i].text == search_keyword :
            elems[i].click()
    except : pass
```

```python
sel_ott = '#mainContent > div > div.box_basic > div.info_detail > div.detail_btn > div > a'
elem_ott = driver.find_elements(By.CSS_SELECTOR, sel_ott)
```

```python
elem_ott[0].text, elem_ott[1].text, elem_ott[2].text, elem_ott[3].text
```

```python
ott_list = []
for j in range(0,len(elem_ott)) :
    ott_list.append(elem_ott[j].text)
```

```python
ott_list
```

```python
base_url = 'https://movie.daum.net/search?q='
search_keyword = title_list

movie_ott_list = []

for n in title_list :
    url = base_url + n
    driver.get(url)
    sel = '#mainContent > div > div.box_searchinfo > div.detail_searchinfo.movie.search_result > ul > li > div > div > strong > a'
    elems = driver.find_elements(By.CSS_SELECTOR, sel)
    for i in range(0,len(elems)) :
        try :
            if elems[i].text == n :
                elems[i].click()
        except : pass
    time.sleep(2)
    sel_ott = '#mainContent > div > div.box_basic > div.info_detail > div.detail_btn > div > a'
    elem_ott = driver.find_elements(By.CSS_SELECTOR, sel_ott)
    ott_list = []
    for j in range(0,len(elem_ott)) :
        ott_list.append(elem_ott[j].text)
    movie_ott_list.append(ott_list)
```

```python
movie_ott_list
```

```python
movie_code['ott'] = movie_ott_list
movie_code.head()
```

```python
movie_code.to_csv('movie_ott.csv', encoding='utf-8-sig')
```
