#### 문제 1

- total_bill 대비 팁 요금의 비율을 계산하고 가공필드 tip_pt 열에 저장하시오.

- 팁의 비율이 요일과 점심/저녁 여부, 인원수에 어떤 영향을 받는지 살펴본다.

- 어떤 요인이 가장 크게 작용하는지 판단할 수 있는 방법이 있는가?

```python
# total_bill 대비 팁 요금의 비율을 계산하고 가공필드 tip_pt 열에 저장하시오.
tips['tip_pt'] = tips['tip']/tips['total_bill']
tips.head()
tips.tail()
```

```python
# 각 요일별 손님의 수를 확인하세요
tips.groupby('day').count() #그룹된 내용의 각컬럼별 빈도를 계산해서 반환 
tips.groupby('day').size() #
```

```python
# 각 요일별 한테이블당 인원수의 평균을 확인하세요
tips.groupby('day')['size'].mean()
```

```python
# 점심/저녁 테이블 수를 비교 하세요
tips.groupby('time').size()
# 점심/저녁 손님 수를 비교 하세요
tips.groupby('time')['size'].sum()
```

```python
# 점심 저녁 별로 테이블당 인원수의 평균을 구하시오.
tips.groupby('time')[['size']].mean()
```

```python
# 각 요일별 점심/저녁 손님의 수를 피봇 테이블로 확인하세요
tips.pivot_table('tip_pt',aggfunc='count',index='day',columns='time')
```

```python
# 각 요일별 테이블당 인원수의 평균과 최대인원 최소인원을 피봇테이블로 확인하시오
tips.pivot_table('size', index='day',aggfunc=['mean', 'max', 'min'])
```

```python
# 각 요일에 따른 점심과 저녁별 테이블당 인원수의 평균과 최대인원 최소인원을 피봇테이블로 확인하시오
tips.pivot_table('size',index=['day','time'], aggfunc=['mean','max','min'])
```

```python
# 각 요일에 따른 점심과 저녁별 테이블당 인원수의 평균과 최대인원 최소인원을 피봇테이블로 확인하시오
pd.options.display.float_format = '{:.2f}'.format
tips.pivot_table('size',index='time', columns='day',aggfunc=['mean','max','min']).fillna(0)
```

```python
# 각 요일별 tip_pt의 평균을 구하시오
tips.groupby('day')[['tip_pt']].mean()
```

```python
# 시간대별 tip_pt의 평균을 확인하시오
pd.options.display.float_format = '{:.4f}'.format
tips.groupby('time')[['tip_pt']].mean()
```

```python
# 각 요일에 따른 time별 tip_pt의 평균을 확인하시오
pd.options.display.float_format = '{:.4f}'.format
tips.groupby(['day','time'])[['tip_pt']].mean().fillna((0))
```

```python
# 각 요일별로 시간대별 테이블의 평균인원수와 팁비율의 평균을 확인하시오
tips.pivot_table(['size','tip_pt'], index = ['day','time'])
```

#### 연습문제 3

- 타이타닉 승객 데이터를 이용하여 다음 분석을 실시하라. 데이터는 다음과 같이 받을 수 있다.
    - titanic = sns.load_dataset("titanic")
    
- 문제 1

- qcut 명령으로 세 개의 나이 그룹을 만든다.

- 성별, 선실, 나이 그룹에 의한 생존율을 데이터프레임으로 계산한다. 
    - 행에는 성별 및 나이 그룹에 대한 다중 인덱스를 사용하고 열에는 선실 인덱스를 사용한다. 
    - 생존률은 해당 그룹의 생존 인원수를 전체 인원수로 나눈 값이다.

- 성별 및 선실에 의한 생존율을 피봇 데이터 형태로 만든다.

```python
# 데이터 로드하기
titanic = sns.load_dataset("titanic")
```

```python
titanic['age_group'] = pd.qcut(titanic['age'],3, labels=['age1','age2','age3'])
```

```python
titanic.head()
```

```python
groups = titanic.groupby(['class','sex','age_group','alive'])
```

```python
groups[['survived']].count()
```

```python
titanic_df = titanic.pivot_table(values='survived',index=['sex','age_group'],
                                 columns=['class','alive'],aggfunc=['count'])
```

```python
titanic_df
```

```python
titanic_df.sum()
```

```python
sum(titanic_df.sum().values)
```

```python
# class 별 생존율
data = {'First' : (titanic_df[('count','First','yes')]/titanic_df['count']['First'].sum(axis=1))*100,
       'Second' : (titanic_df[('count','Second','yes')]/titanic_df['count']['Second'].sum(axis=1))*100,
       'Third' : (titanic_df[('count','Third','yes')]/titanic_df['count']['Third'].sum(axis=1))*100}
```

```python
titanic_survived_rate = pd.DataFrame(data, columns=['First','Second','Third'])
titanic_survived_rate
```
