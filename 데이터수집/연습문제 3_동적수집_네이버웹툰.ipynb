```python
# 패키지 임포트
import selenium
from selenium import webdriver
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By # 셀레니움 4.0부터 포함된 함수(필수)
import time
```

```python
# 데이터 하나를 df로 생성 후 list에 저장 
# 전체 검색결과를 담을 list
result=[]
```

```python
# 1. webdriver 객체 생성
driver = webdriver.Chrome('./driver/chromedriver')
```

```python
url = 'https://comic.naver.com/genre/bestChallenge.nhn'
driver.get(url)
```

```python
sel ='#content > div:nth-child(2) > ul > li > div'
```

```python
elem = driver.find_elements(By.CSS_SELECTOR,sel)
```

```python
len(elem)
```

```python
for e in elem :
    res = e.text.split('\n')
    if len(res) == 7:
        result.append(pd.DataFrame({'제목':res[1],'작가':res[2],'별점':res[4],
                                    '조회수':res[6]},index = range(0,1)))
    else : 
        result.append(pd.DataFrame({'제목':res[0],'작가':res[1],'별점':res[3],
                                        '조회수':res[5]},index = range(0,1)))
```

```python
pd.concat(result).reset_index(drop=True)
```

```python
driver.close()
```

20페이지까지

```python
result = []
```

```python
# 1. webdriver 객체 생성
driver = webdriver.Chrome('./driver/chromedriver')
```

```python
for i in range(1,21) : 
    url = 'https://comic.naver.com/genre/bestChallenge.nhn?page=' + str(i)
    driver.get(url)
    time.sleep(2)
    sel = '#content > div:nth-child(2) > ul > li > div'
    elem = driver.find_elements(By.CSS_SELECTOR,sel)
    for e in elem :
        res = e.text.split('\n')
        if len(res) == 7 :
            result.append(pd.DataFrame({'제목':res[1],'작가':res[2],'별점':res[4],'조회수':res[6]},index=range(0,1)))
        else :
            result.append(pd.DataFrame({'제목':res[0],'작가':res[1],'별점':res[3],
                                        '조회수':res[5]},index=range(0,1)))
```

```python
pd.concat(result).reset_index(drop=True).to_csv('./crawl_data/webtoon.csv')
```
