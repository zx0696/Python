```python
# 영화명(한국어) 검색 및 ott 수집
base_url = 'https://movie.daum.net/search?q='
search_keyword = title_list_ko

movie_ott_list_ko = []

for n in search_keyword :
    url = base_url + n
    driver.get(url)
    sel = '#mainContent > div > div.box_searchinfo > div.detail_searchinfo.movie.search_result > ul > li > div > div > strong > a'
    elems = driver.find_elements(By.CSS_SELECTOR, sel)
    for i in range(0,len(elems)) :
        try :
            if elems[i].text == n :
                elems[i].click()
        except : pass
    time.sleep(2)
    sel_ott = '#mainContent > div > div.box_basic > div.info_detail > div.detail_btn > div > a'
    elem_ott = driver.find_elements(By.CSS_SELECTOR, sel_ott)
    ott_list = []
    for j in range(0,len(elem_ott)) :
        ott_list.append(elem_ott[j].text)
    movie_ott_list_ko.append(ott_list)
```

```python
# 영화명(영어) 검색 및 ott 수집
base_url = 'https://movie.daum.net/search?q='
search_keyword = title_list_en

movie_ott_list_en = pd.DataFrame(columns=['Title','openDt','watchGd','grade'])

for n in search_keyword :
    url = base_url + str(n)
    driver.get(url)
    time.sleep(1)
    sel = '#mainContent > div > div.box_searchinfo > div.detail_searchinfo.movie.search_result > ul > li > div > div > strong > a'
    elems = driver.find_elements(By.CSS_SELECTOR, sel)
    sel_en = '#mainContent > div > div.box_searchinfo > div.detail_searchinfo.movie.search_result > ul > li > div > div > span'
    elems_en = driver.find_elements(By.CSS_SELECTOR, sel_en)
    for i in range(0,len(elems_en)) :
        try :
            if elems_en[i].text == n :
                elems[i].click()
        except : pass
    time.sleep(2)
    sel_info = '#mainContent > div > div.box_basic > div.info_detail > div.detail_cont > div > dl'
    elems_info = driver.find_elements(By.CSS_SELECTOR, sel_info)
#     sel_tit = '#mainContent > div > div.box_basic > div.info_detail > div.detail_tit > h3 > span.txt_tit'
#     elem_tit = driver.find_element(By.CSS_SELECTOR, sel_tit)
    title = n
    for i in range(0,len(elems_info)) :
        if elems_info[i].text.startswith('개봉') == True :
            openDt = elems_info[i].text
        elif elems_info[i].text.startswith('등급') == True :
            watchGd = elems_info[i].text
        elif elems_info[i].text.startswith('평점') == True :
            grade = elems_info[i].text
    movie_ott_list_en = movie_ott_list_en.append({'Title':title,'openDt':openDt,'watchGd':watchGd,'grade':grade},ignore_index=True)
```
