```python
import selenium
from selenium import webdriver
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By # 셀레니움 4.0부터 포함된 함수(필수)
import time
```

```python
driver = webdriver.Chrome('./데이터수집/driver/chromedriver.exe')
```

```python
url = 'https://www.megabox.co.kr/movie-detail/comment?rpstMovieNo=21049700'
driver.get(url)
```

```python
story_point = '#contentData > div > div.movie-idv-story > ul > li > div.story-area > div.story-box > div > div.story-cont > div.story-point'
story_recomm = '#contentData > div > div.movie-idv-story > ul > li > div.story-area > div.story-box > div > div.story-cont > div.story-recommend'
story_txt = '#contentData > div > div.movie-idv-story > ul > li > div.story-area > div.story-box > div > div.story-cont > div.story-txt'
```

```python
elem_point = driver.find_elements(By.CSS_SELECTOR,story_point)
elem_recomm = driver.find_elements(By.CSS_SELECTOR,story_recomm)
elem_txt = driver.find_elements(By.CSS_SELECTOR,story_txt)
```

```python
result = []
```

```python
for i in range(10) :
    url = 'https://www.megabox.co.kr/movie-detail/comment?rpstMovieNo=21049700'
    driver.get(url)
    time.sleep(2)
    sel_page = '#contentData > div > div.movie-idv-story > nav.pagination > *'
    elem = driver.find_elements(By.CSS_SELECTOR,sel_page)[0:10]
    elem[i].click()
    time.sleep(2)
    
    story_point = '#contentData > div > div.movie-idv-story > ul > li > div.story-area > div.story-box > div > div.story-cont > div.story-point'
    story_recomm = '#contentData > div > div.movie-idv-story > ul > li > div.story-area > div.story-box > div > div.story-cont > div.story-recommend'
    story_txt = '#contentData > div > div.movie-idv-story > ul > li > div.story-area > div.story-box > div > div.story-cont > div.story-txt'
    
    elem_point = driver.find_elements(By.CSS_SELECTOR,story_point)
    elem_recomm = driver.find_elements(By.CSS_SELECTOR,story_recomm)
    elem_txt = driver.find_elements(By.CSS_SELECTOR,story_txt)
    
    res_point = [e_p.text for e_p in elem_point]
    res_recomm = [e_r.text for e_r in elem_recomm]
    res_txt = [e_t.text for e_t in elem_txt]
    
    result.append(pd.DataFrame({'평점':res_point,'추천인':res_recomm,'관람평':res_txt}))
```

```python
pd.concat(result).reset_index(drop=True).to_csv('./movie.csv',encoding = 'utf-8-sig')
```
