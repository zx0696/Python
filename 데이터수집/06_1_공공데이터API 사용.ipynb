```python
 QT = '1' # 월요일에 여는 약국
QN = quote('삼성약국')# 기관명(약국이름-한글로전달)
Q0 = quote('서울특별시')#광역시도(한글명전달)
Q1 = quote('강남구') # 시군구
ORD = 'NAME' #정렬기준
pageNo='1'
numOfRows='6000'
```

```python
paramset = "serviceKey="+serviceKey+"&"\
            + "Q0="+Q0 + "&"\
            + "ORD="+ ORD + "&" \
            + "numOfRows=" + numOfRows
```

```python
url = endpoint + paramset
result = requests.get(url)
bs_obj = BeautifulSoup(result.content,'html.parser')
bs_obj
```

```python
res = bs_obj.findAll('item')
```

```python
len(res)
```

```python
res[1].find('dutyname').text
```

```python
# 서울시 약국 정보  df
import pandas as pd
name = [r.dutyname.text for r in res]
addr = [r.dutyaddr.text for r in res]
tel  = [r.dutytel1.text for r in res]
```

```python
len(name),len(addr),len(tel)
```

```python
df = pd.DataFrame({'약국명':name,'주소':addr,'전화번호':tel})
df.tail()
```

```python
df.to_csv('crawl_data/서울시약국정보.csv')
```

### 서울시 약국 중 월요일 밤 9시 이후까지 운영하는 약국 정보

```python
int(res[0].dutytime1c.text) > 2100
```

```python
# 빈 리스트 생성
name = []
adr = []
tel = []
count = 0
```

```python
for item in res : 
    t_item = item.find('dutytime1c')
    if t_item != None : 
        c_time = int(t_item.text)
        if c_time > 2100 :  
            count += 1
            name.append(item.find('dutyname').text)
            adr.append(item.find('dutyaddr').text)
            tel.append(item.find('dutytel1').text)
print(count)
```

```python
for item in res : 
    t_item = item.find('dutytime1c')
    if t_item != None : 
        c_time = int(t_item.text)
        if c_time > 2100 : 
            try : 
                count += 1
                name.append(item.find('dutyname').text)
                adr.append(item.find('dutyaddr').text)
                tel.append(item.find('dutytel1').text)
            except :
                tel.append('no data')
print(count)
```

```python
df_mon_night = pd.DataFrame({'약국명':name,'주소':adr,'전화번호':tel})
```

```python
df_mon_night.to_csv('./crawl_data/월요일심야영업약국정보.csv')
```
